## GPS Data

### 五一假期厦门巡游车辆GPS数据

```python
xlsx_gps = pd.read_excel(data_path_org+'五一假期厦门巡游车辆GPS数据.xlsx')
```

```python
for n in xlsx_gps.columns:
    print(n,'/',xlsx_gps.loc[0][n])
```

```python
xlsx_gps.to_csv(data_path+'GPS_Touring.csv', index=False, encoding='utf_8_sig')
```
