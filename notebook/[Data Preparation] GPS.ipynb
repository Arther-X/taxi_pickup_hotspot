```python
import pandas as pd
import datetime

DATA_PATH = '/data/'
DATA_TAXI_GPS = DATA_PATH+'taxiGps/'
# DATA_FILE_NAME = 'taxiGps20180501-00-12.csv'
DATA_FILE_NAME = 'taxiGps20180501-12-24.csv'


df = pd.read_csv(DATA_TAXI_GPS+DATA_FILE_NAME)
df.dropna(subset=['GETON_LONGITUDE', 
                  'GETON_LATITUDE', 
                  'GETOFF_LONGITUDE', 
                  'GETOFF_LATITUDE',
                  'PASS_MILE',
                  'CAR_NO'], inplace=True)



df['GPS_TIME'] = pd.to_datetime(df['GPS_TIME'])

t_df = df
t_df['GPS_TIME'] = t_df['GPS_TIME'].apply(
    lambda dt: datetime.datetime(dt.year, dt.month, dt.day, dt.hour)
)
print(t_df.groupby('GPS_TIME')['LONGITUDE'].count())

```
